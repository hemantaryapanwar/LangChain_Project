```python
import os 
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model=init_chat_model(model="gpt-4.1")
model
```

```python
response=model.invoke("Hello, how are you?")
response.content
```

### ChatOpenAI (direct import)

Use `ChatOpenAI` from `langchain_openai` when you want to configure the OpenAI model directly.

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4.1")
llm
```

```python
response = llm.invoke("Hello, how are you?")
response.content
```

Streaming and Batch

```python
model.invoke("write essay of 200 words on AI")
```

```python
for chunk in model.stream("write essay of 200 words paragraph on AI"):
    print(chunk.text, end="", flush=True)
```

Batch


```python
responses=model.batch(["write essay of 50 words on AI", "write essay of 200 words on AI",
"what is quantum computing?",
"what is the weather in Tokyo?"])

for response in responses:
    print(response.content )




```
