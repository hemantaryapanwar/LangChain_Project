```python
import os
from langchain.chat_models import init_chat_model
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")

model=init_chat_model(model="gpt-4.1")
model
response=model.invoke("Hello, how are you?")
response.content
```

```python
from langchain.tools import tool

@tool
def get_weather(city: str)->str:
    """Get the weather in a given city"""
    return f"The weather in {city} is sunny"


model_with_tools=model.bind_tools([get_weather])
response=model_with_tools.invoke("What is the weather in Tokyo?")
print(response)
for tool_call in response.tool_calls:
    print(f"tool_call: {tool_call['name']}")
    print(f"args: {tool_call['args']}") 

```

### Tool Execution


```python
messages=[{"role":"user", "content":"what is weather in Boston"}]
ai_message=model_with_tools.invoke(messages)
messages.append(ai_message)

#Step 2 Execute Tools
for tool_call in ai_message.tool_calls:
    total_result=get_weather.invoke(tool_call)
    messages.append(total_result)

final_response = model_with_tools.invoke(messages)
print(final_response.text)

```
